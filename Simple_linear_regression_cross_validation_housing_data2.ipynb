```python
# Import required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data
url=("https://raw.githubusercontent.com/DrSaadLa/MLLabs/main/data/housing.csv")
```

```python
# Read the data
data=pd.read_csv(url)
```

```python
# check the first 10 observations
data.head(10)
```

```python
# check the last 10 observations
data.tail(10)
```

```python
# check data info
data.info()
```

```python
# Descriptive statistics
data.describe().T
```

```python
# check correlations (feature selection)
sns.set(rc={"figure.figsize": (10,8)})
sns.heatmap(data.corr(), cmap="RdYlBu", square=True)
```

```python
# define target
y=data["Price"]
```

```python
# define x
x=data["Avg. Area Number of Rooms"]
```

```python
# check type y and x
type(y)
type(x)
```

```python
# reshape y to 2D
y_reshaped=np.array(y).reshape(-1,1)
```

```python
# reshape x to 2D
x_reshaped=np.array(x).reshape(-1,1)
```

```python
# observe the scatter between x and y
plt.scatter(x_reshaped, y_reshaped, color="blue", alpha=0.5)
plt.xlabel("Avg. Area Number of Rooms")
plt.ylabel("Price")
plt.show()
```

```python
# add the regression line to scatterplot
sns.lmplot(x="Avg. Area Number of Rooms", y="Price", data=data)
```

```python
# import linear regression
from sklearn.linear_model import LinearRegression
```

```python
# create linear regression object
lreg2=LinearRegression()
```

```python
# import cross validation score
from sklearn.model_selection import cross_val_score
```

```python
# use 5-Kfold
cv_results=cross_val_score(lreg2, x_reshaped, y_reshaped, cv=5)
cv_results
```

```python
print("The mean coefficient of determination is:", cv_results.mean())
```

```python
# make another metric score
from sklearn.metrics import make_scorer
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mae=make_scorer(mean_absolute_error)
```

```python
cv_results2=cross_val_score(lreg2, x_reshaped, y_reshaped, cv=5, scoring=mae)
```

```python
cv_results2
```

```python
print("The mean mae is:", cv_results2.mean())
```

```python
mse=make_scorer(mean_squared_error)
```

```python
cv_results3=cross_val_score(lreg2, x_reshaped, y_reshaped, cv=5, scoring=mse)
```

```python
print("The mean mse is:", cv_results3.mean())
```
