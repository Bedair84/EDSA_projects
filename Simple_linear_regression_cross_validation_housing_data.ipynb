```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
kf=KFold(n_splits=5, shuffle=True, random_state=10123)
```

```python
splits=kf.split(x)
```

```python
for train_index, val_index in splits:
    print(len(train_index), len(val_index))
```

```python
lreg=LinearRegression()
```

```python
for train_index, val_index in splits:
    x_train, y_train = x[train_index], y[train_index]
    x_val, y_val = x[val_index], y[val_index]
    lreg.fit(x_train, y_train)
    preds = lreg.predict(x_val)
    print("The MSE metric: {:.5 f}".format(mean_squared_error(y_val, preds)))
    print("The MAE metric: {:.5 f}".format(mean_absolute_error(y_val, preds)))
    print()
```
